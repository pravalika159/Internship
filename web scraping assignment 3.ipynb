### 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
product_urls=[]
start=1
end=4
for page in range(start,end):
    urls=amazon.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in urls:
        product_urls.append(i.get_attribute('href'))
    next_button=amazon.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[22]/div/div/span/a[3]')
    next_button.click()
    time.sleep(5)
```

```python
len(product_urls)
```

```python
Brand=[]
Product_Name=[]
Price=[]
With_Exchange=[]
Expected_Delivery=[]
Availability=[]
Product_Url=[]
for url in product_urls:
    amazon.get(url)
    time.sleep(3)
    try:
        brand=amazon.find_element(By.XPATH,'//tr[@class="a-spacing-small po-brand"]/td[2]')
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')
    try:
        product=amazon.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        Product_Name.append(product.text)
    except NoSuchElementException:
        Product_Name.append('-')
    try:
        price=amazon.find_element(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
    try:
        exchange=amazon.find_element(By.XPATH,'//div[@class="a-section a-spacing-none a-padding-none show"]')
        With_Exchange.append(exchange.text)
    except NoSuchElementException:
        With_Exchange.append('-')
    try:
        delivery=amazon.find_element(By.XPATH,'//div[@class="a-spacing-base"][1]')
        Expected_Delivery.append(delivery.text)
    except NoSuchElementException:
        Expected_Delivery.append('-')
    try:
        availability=amazon.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        Availability.append(availability.text)
    except NoSuchElementException:
        Availability.append('-')
    try:
        Product_Url.append(url)
    except NoSuchElementException:
        Product_Url.append('-')
        
```

```python
df=pd.DataFrame({'BRAND':Brand,'PRODUCT NAME':Product_Name,'PRICE':Price,'WITH EXCHANGE':With_Exchange,
                 'EXPECTED DELIVERY':Expected_Delivery,'AVAILABILITY':Availability,'PRODUCT URL':Product_Url})
df
```

```python
csv_file=df.to_csv("ipad_details.csv")
csv_file
```

```python
Hostel_Names=[]
names=hostel.find_elements(By.XPATH,'//div[@class="property-info-header"]/div[2]')
for i in names:
    Hostel_Names.append(i.text)
Distance=[]
dist=hostel.find_elements(By.XPATH,'//span[@class="distance-description"]')
for i in dist:
    Distance.append(i.text)
Ratings=[]
ratings=hostel.find_elements(By.XPATH,'//div[@class="property-info-header"]/div/div/div/span[1]')
for i in ratings:
    Ratings.append(i.text)
Reviews=[]
reviews=hostel.find_elements(By.XPATH,'//div[@class="rating medium theme-null"]/div/span[2]')
for i in reviews:
    Reviews.append(i.text)
Overall_Reviews=[]
all_reviews=hostel.find_elements(By.XPATH,'//span[@class="left-margin"]')
for i in all_reviews:
    Overall_Reviews.append(i.text)
Private_From=[]
private=hostel.find_elements(By.XPATH,'//div[@class="property-accommodation-prices"]/div[1]')
for i in private:
    Private_From.append(i.text)
Dorm_From=[]
dorm=hostel.find_elements(By.XPATH,'//div[@class="property-accommodation-prices"]/div[2]')
for i in dorm:
    Dorm_From.append(i.text)
```
