### 1) Write a python program to display all the header tags from wikipedia.org.

```python
page=requests.get("https://www.wikipedia.org/")
```

```python
page
```

```python
data1=BeautifulSoup(page.content)
```

```python
data1
```

### HEADERS:

```python
headers=[]
for i in data1.find_all('div',class_="central-featured-lang"):
    headers.append(i.text)
headers
```

```python
print(len(headers))
```

### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
page=requests.get("https://www.imdb.com/list/ls091520106/")
```

```python
page
```

```python
data2=BeautifulSoup(page.content)
```

```python
data2
```

```python
names=[]
for i in data2.find_all('h3',class_="lister-item-header"):
    names.append(i.text)
names
```

```python
print(len(names))
```

```python
ratings=[]
for i in data2.find_all('div',class_="ipl-rating-star small"):
    ratings.append(i.text)
ratings
```

```python
print(len(ratings))
```

```python
year_of_release=[]
for i in data2.find_all('span',class_="lister-item-year text-muted unbold"):
    year_of_release.append(i.text)
year_of_release
```

```python
print(len(year_of_release))
```

```python
#creating dataframe
import pandas as pd
df=pd.DataFrame({'NAMES':names,'RATINGS':ratings,"YEAR OF RELEASE":year_of_release})
df
```
