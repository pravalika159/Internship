### 1) Write a python program to display all the header tags from wikipedia.org.

```python
page=requests.get("https://www.wikipedia.org/")
```

```python
page
```

```python
data1=BeautifulSoup(page.content)
```

```python
data1
```

### HEADERS:

```python
headers=[]
for i in data1.find_all('div',class_="central-featured-lang"):
    heading=i.a.strong.text
    headers.append(heading)
print(headers)
```

```python
print(len(headers))
```

### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
page=requests.get("https://www.imdb.com/list/ls091520106/")
```

```python
page
```

```python
data2=BeautifulSoup(page.content)
```

```python
data2.text
```

```python
names=[]
for i in data2.find_all('h3',class_="lister-item-header"):
    name=i.a.text
    names.append(name)
print(names)
```

```python
print(len(names))
```

```python
ratings=[]
for i in data2.find_all('div',class_="ipl-rating-star small"):
    ratings.append(i.text.strip())
ratings
```

```python
print(len(ratings))
```

```python
year_of_release=[]
for i in data2.find_all('span',class_="lister-item-year text-muted unbold"):
    year_of_release.append(i.text)
year_of_release
```

```python
print(len(year_of_release))
```

```python
#creating dataframe
import pandas as pd
df=pd.DataFrame({'NAMES':names,'RATINGS':ratings,"YEAR OF RELEASE":year_of_release})
df
```

### 3) Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL.

```python
page=requests.get("https://www.dineout.co.in/")
```

```python
page
```

```python
data3=BeautifulSoup(page.content)
```

```python
data3
```

```python
NAMES=[]
for i in data3.find_all('div',class_="restnt-name ellipsis"):
    names.append(i.text)
print(NAMES) 
```

```python
cuisine=[]
for i in data3.find_all('div',class_="detail-info"):
    x=i.a.text
    cuisine.append(x)
print(cuisine)
```

```python
location=[]
for i in data3.find_all('div',class_="restnt-loc ellipsis"):
    y=i.a.a.text
    location.append(y)
print(location)
```

```python
ratings=[]
for i in data3.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(i.text)
print(ratings)
```

```python
image=[]
for i in data3.find_all("div",class_="img cursor"):
    image.append(i['data-src'])
image
```

### 4) Write s python program to display list of respected former finance minister of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
page=requests.get("https://presidentofindia.nic.in/former-presidents")
```

```python
page
```

```python
data4=BeautifulSoup(page.content)
```

```python
data4
```

```python
names=[]
for i in data4.find_all('div',class_="desc-sec"):
    names.append(i.text.strip())
names
```

```python
print(len(names))
```

```python
office=[]
for i in data4.find_all("div",class_="desc-sec"):
    office.append(i.h5.text)
print(office)
```

```python
import pandas as pd
df=pd.DataFrame({'NAMES':names,'OFFICE':office})
df
```
